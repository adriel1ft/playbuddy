```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

import os
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv())

client_id = os.getenv("SPOTIFY_CLIENT_ID")
client_secret = os.getenv("SPOTIFY_CLIENT_SECRET")
redirect_uri = os.getenv("SPOTIFY_REDIRECT_URI")
access_token = os.getenv("SPOTIFY_ACCESS_TOKEN")

```

```python
auth_manager = SpotifyClientCredentials(client_id=client_id, 
                                        client_secret=client_secret)
sp = spotipy.Spotify(auth_manager=auth_manager)

```

```python
results = sp.search(q='Gal Costa', type='artist')
artist_id = results['artists']['items'][0]['id']
```

```python
top_track_results = sp.artist_top_tracks(artist_id,  available_markets='BR')
for track in top_track_results['tracks'][:5]:
    print('Track: ', track['name'])
    print('Album: ', track['album']['name'])
    print('Artist: ', track['album']['artists'][0]['name'])
    print('Popularity: ', track['popularity'])
    print('--------------------------------')
```

```python
trid = artist_id  # or extract from URI if needed
top_track_results = sp._get(f"artists/{trid}/top-tracks", market="BR")
```

```python
results = sp.search(q='artist:Gal Costa', type='track', limit=10, market='BR')
for track in results['tracks']['items']:
    minutes = track['duration_ms'] // 60000
    seconds = (track['duration_ms'] % 60000) // 1000
    print(f"{track['name']} ({minutes}:{seconds:02d}) - Album: {track['album']['name']}")
```

```python
# First, get an album ID (e.g., from a search result)
album_id = results['tracks']['items'][0]['album']['id']

# Then get all tracks from that album
album_tracks = sp.album_tracks(album_id)
for i, track in enumerate(album_tracks['items'], 1):
    minutes = track['duration_ms'] // 60000
    seconds = (track['duration_ms'] % 60000) // 1000
    print(f"{i}. {track['name']} ({minutes}:{seconds:02d})")
```

```python
albums = sp.artist_albums(artist_id, limit=10)
for album in albums['items']:
    print(f"{album['name']} ({album['release_date']})")
```

```python
all_albums = []
offset = 0
limit = 20

while True:
    results = sp.artist_albums(artist_id, include_groups='album', limit=limit, offset=offset)
    all_albums.extend(results['items'])
    if results['next'] is None:
        break
    offset += limit

print(f"Total albums: {len(all_albums)}\n")
for album in all_albums:
    print(f"{album['name']} ({album['release_date']})")
```
